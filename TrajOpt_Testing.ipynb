```python
import numpy as np
from scipy.optimize import minimize

# Example objective function (Rosenbrock function)
def rosenbrock(x):
    return sum(100.0 * (x[:, 1:] - x[:, :-1]**2)**2 + (1 - x[:, :-1])**2)

# Main optimization function
def main_optimization():
    # Define the initial guess x0 with shape (3, 4)
    x0 = np.zeros((3, 4))

    # Perform optimization using the minimize function
    result = minimize(fun=rosenbrock,
                      x0=x0,
                      method='SLSQP')

    return result

# Example usage
result = main_optimization()

# Print the result
print("Optimal solution:")
print(result.x)
print("Optimal function value:")
print(result.fun)
```
